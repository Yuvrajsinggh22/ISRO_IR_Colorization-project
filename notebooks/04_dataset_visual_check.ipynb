```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import random

PROCESSED_DIR = Path("../data/processed")
split = "train"

input_dir = PROCESSED_DIR / split / "inputs"
target_dir = PROCESSED_DIR / split / "targets"

input_files = sorted(input_dir.glob("*.npy"))
target_files = sorted(target_dir.glob("*.npy"))

print("Inputs:", len(input_files))
print("Targets:", len(target_files))
```

```python
idx = random.randint(0, len(input_files) - 1)

inp = np.load(input_files[idx])
tgt = np.load(target_files[idx])

print("File:", input_files[idx].name)
print("Input shape:", inp.shape, inp.min(), inp.max())
print("Target shape:", tgt.shape, tgt.min(), tgt.max())
```

```python
titles = [
    "Thermal B10",
    "NDVI",
    "NDWI",
    "NDBI",
    "SWIR2 B7",
]

plt.figure(figsize=(18, 6))

for i in range(5):
    plt.subplot(1, 6, i + 1)
    plt.imshow(inp[:, :, i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

plt.subplot(1, 6, 6)
plt.imshow(tgt)
plt.title("Target RGB")
plt.axis("off")

plt.tight_layout()
plt.show()
```
